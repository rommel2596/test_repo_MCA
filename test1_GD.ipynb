```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Definir funcion
def f_sphere(x, y):
    return x**2 + y**2

```

```python
# Define a meshgrid for plotting the function

x_range = np.arange(-15,15, 0.1)
y_range = np.arange(-15,15, 0.1)
X, Y = np.meshgrid(x_range,y_range)
Z = f_sphere(X,Y)
```

```python
# Plot the function

fig = plt.figure()

# Create axis
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X,Y,Z, cmap='coolwarm', alpha=0.5) #Se agrega alpha para la transparencia

# Para la tarea :D
ax.scatter(0,0,0, c='r', marker='o')

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('function')

plt.show()
```
